```python
grouped_data = df.groupby('study_load')['academic_performance'].mean()

# Print the average academic performance for each study load
for study_load, avg_academic_performance in grouped_data.items():
    print(f'Study Load {study_load}: Average Academic Performance {avg_academic_performance:.2f}')
```

```python
# Filter rows where study_load is 5
filtered_df = df[df['study_load'] == 5]

# Calculate the average academic_performance for those rows
average_academic_performance = filtered_df['academic_performance'].mean()

# Print the result
print(f"The average academic performance for those who rated a study load of 5 is: {average_academic_performance}")
```
