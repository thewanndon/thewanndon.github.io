```python
for column in df.columns:
    # Print min and max for each column
    print(f"{column}:")
    print(f"  Min: {df[column].min()}")
    print(f"  Max: {df[column].max()}")
    print()
```

```python
grouped_data = df.groupby('study_load')['academic_performance'].mean()

# Print the average academic performance for each study load
for study_load, avg_academic_performance in grouped_data.items():
    print(f'Study Load {study_load}: Average Academic Performance {avg_academic_performance:.2f}')
```

```python
# Filter rows where study_load is 5
filtered_df = df[df['study_load'] == 5]

# Calculate the average academic_performance for those rows
average_academic_performance = filtered_df['academic_performance'].mean()

# Print the result
print(f"The average academic performance for those who rated a study load of 5 is: {average_academic_performance}")
```

```python
grouped_data_esteem = df.groupby('study_load')['self_esteem'].mean()

# Print the average academic performance for each study load
for study_load, avg_self_esteem in grouped_data_esteem.items():
    print(f'Study Load {study_load}: Average Self Esteem {avg_self_esteem:.2f}')
```

```python
grouped_data_depression = df.groupby('study_load')['depression'].mean()

# Print the average academic performance for each study load
for study_load, avg_depression in grouped_data_depression.items():
    print(f'Study Load {study_load}: Average Depression Level {avg_depression:.2f}')
```

```python
grouped_data_anxiety = df.groupby('study_load')['anxiety_level'].mean()

# Print the average academic performance for each study load
for study_load, avg_anxiety in grouped_data_anxiety.items():
    print(f'Study Load {study_load}: Average Anxiety Level {avg_anxiety:.2f}')
```
